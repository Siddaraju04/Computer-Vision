```python
# Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define a function to apply Gabor filter
def apply_gabor_filter(image, kernel_size, sigma, theta, lamda, gamma, psi=0):
    kernel = cv2.getGaborKernel((kernel_size, kernel_size), sigma, theta, lamda, gamma, psi, ktype=cv2.CV_32F)
    filtered_image = cv2.filter2D(image, cv2.CV_8UC3, kernel)
    return filtered_image, kernel

# Load the image in grayscale
image_path = r'C:\Users\Siddaraju.B.R\Downloads\daylight-forest-glossy-443446.jpg'  # Replace with the path to your image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Display the original image
plt.figure(figsize=(10, 10))
plt.subplot(3, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Parameters for Gabor filter
kernel_size = 31  # Size of Gabor kernel
sigma = 4.0       # Standard deviation of Gaussian function
lamda = 10.0      # Wavelength of sinusoidal function
gamma = 0.5       # Spatial aspect ratio
angles = [0, np.pi/4, np.pi/2, 3*np.pi/4]  # Orientations (in radians)

# Apply Gabor filters at different orientations
for i, theta in enumerate(angles):
    filtered_image, kernel = apply_gabor_filter(image, kernel_size, sigma, theta, lamda, gamma)
    plt.subplot(3, 3, i + 2)
    plt.imshow(filtered_image, cmap='gray')
    plt.title(f'Gabor: θ={theta:.2f}')
    plt.axis('off')

# Show all images
plt.tight_layout()
plt.show()
```
