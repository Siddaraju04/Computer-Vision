```python
# Import necessary libraries
import cv2
import matplotlib.pyplot as plt

# Load the image in grayscale
image_path = r'C:\Users\Siddaraju.B.R\Downloads\OIP.jpeg'# Replace with the path to your image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Display the original image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Apply Gaussian Blur to reduce noise
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

# Apply Laplacian operator
laplacian = cv2.Laplacian(blurred_image, cv2.CV_64F)

# Convert the result to 8-bit absolute value
laplacian_edges = cv2.convertScaleAbs(laplacian)

# Display the edges
plt.subplot(1, 2, 2)
plt.imshow(laplacian_edges, cmap='gray')
plt.title('Laplacian Edge Detection')
plt.axis('off')

plt.tight_layout()
plt.show()

# Save the edge-detected image (optional)
output_path = 'laplacian_edges.jpg'
cv2.imwrite(output_path, laplacian_edges)
```
