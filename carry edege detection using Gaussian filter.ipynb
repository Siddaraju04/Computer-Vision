```python
# Import necessary libraries
import cv2
import matplotlib.pyplot as plt

# Load the image in grayscale
image_path = r'C:\Users\Siddaraju.B.R\Downloads\daylight-forest-glossy-443446.jpg'  # Replace with your image file path
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply Gaussian filter to reduce noise
gaussian_blurred = cv2.GaussianBlur(image, (5, 5), 1.4)  # (5, 5) is the kernel size, 1.4 is the standard deviation

# Perform edge detection using Canny algorithm
edges = cv2.Canny(gaussian_blurred, 100, 200)  # Adjust thresholds (100, 200) as needed

# Display the results
plt.figure(figsize=(12, 6))

# Original image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Gaussian blurred image
plt.subplot(1, 3, 2)
plt.imshow(gaussian_blurred, cmap='gray')
plt.title('Gaussian Blurred Image')
plt.axis('off')

# Edge-detected image
plt.subplot(1, 3, 3)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection with Canny')
plt.axis('off')

plt.tight_layout()
plt.show()
```
